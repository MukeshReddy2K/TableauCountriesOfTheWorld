# Renaming the columns with more readability

```python
# Renaming columns
df.columns = ['Country', 'Region', 'Population', 'Area', 'Pop_Density', 'Coastline_Ratio', 'Net_Migration', 'Infant_Mortality', 'GDP_Per_Capita', 'Literacy_percentage', 'Phones', 'Arable_percentage', 'Crops_percentage', 'Other_percentage', 'Climate', 'Birthrate', 'Deathrate', 'Agriculture', 'Industry', 'Service']

# Display the DataFrame with updated column names
df.head()
```

# Preprocessing the Dataset

## Replacing the values ',' with '.'

```python
df.replace(',', '.', regex=True, inplace=True)
```

```python
df.head()
```

## Checking the null values for every column

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
col = df.columns
num_col = df.columns[4:]
```

```python
df.info()
```

## Converting the datatype from object to float

```python
df.iloc[:,4:] = df.iloc[:,4:].astype('float').round(2)
```

```python
df.head()
```

## Filling the Missing values according to the mean of the specific region

```python
# Taking the mean according to 'Region' and replacing with mean value
for i in num_col:
  df[i] = (df[i].fillna(df.groupby('Region')[i].transform('mean')).astype(float)).round(2)
```

```python
df.head()
```

# Checking if any null values remaining

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.iloc[:,8] = df.iloc[:,8].astype('int')
df.iloc[:,14] = df.iloc[:,14].astype('int')
```
